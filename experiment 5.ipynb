```python
#Image Rotation
from PIL import Image
import numpy as np
from scipy import ndimage
from matplotlib import pyplot as plt
%matplotlib inline
a=Image.open('h.jpg')
b=np.array(a)
c=ndimage.rotate(b,90,cval=110)
d=Image.fromarray(c)
d.save('b.jpg') 
plt.imshow(d)
```

```python
#Image scaling
from PIL import Image
import numpy as np
from scipy import ndimage
from matplotlib import pyplot as plt
%matplotlib inline


a=Image.open('h.jpg')
b=np.array(a)
c=ndimage.zoom(b,(.5,1,1))
d=Image.fromarray(c)
d.save('b.jpg')
plt.imshow(d)
```
